```python
iris_dataset = load_iris(as_frame = True).data
x_variables = iris_dataset.iloc[:, 0:2]
y_variables = iris_dataset.iloc[:, 3]
```

```python
plt.figure(figsize=(10,8))
sb.barplot(x=list(x_variables.iloc[:, 0]), y=list(y_variables))
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Width(cm)')
plt.title('Sepal Length vs. Petal Width')
```

```python
sb.scatterplot(x=list(x_variables.iloc[:, 0]), y=list(y_variables))
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Width(cm)')
plt.title('Sepal Length vs. Petal Width')
```

```python
lt5 = [datapoint for datapoint in x_variables.iloc[:, 0] if datapoint < 5]
lt6 = [datapoint for datapoint in x_variables.iloc[:, 0] if datapoint < 6 and datapoint >= 5]
lt7 = [datapoint for datapoint in x_variables.iloc[:, 0] if datapoint < 7 and datapoint >= 6]
lt8 = [datapoint for datapoint in x_variables.iloc[:, 0] if datapoint < 8 and datapoint >= 7]
df = pd.DataFrame()
df['category'] = ['less than 5cm', 'less than 6cm', 'less than 7 cm', 'less than 8 cm']
df['number'] = [len(lt5), len(lt6), len(lt7), len(lt8)]

pie, ax = plt.subplots(figsize=[10,6])
plt.pie(x=df.number, autopct="%.1f%%", explode=[0.05]*4, labels=df.category, pctdistance=0.5)
plt.title("Sepal Lengths", fontsize=14);
```
