```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

gender_dataset = pd.read_csv('titanic/gender_submission.csv')
test_dataset = pd.read_csv('titanic/test.csv')
train_dataset = pd.read_csv('titanic/train.csv')
```

```python
train_dataset.head()
```

```python
train_dataset['FareQCut'] = pd.qcut(train_dataset['Fare'], 4)
train_dataset[['FareQCut', 'Survived']].groupby(['FareQCut'], as_index=False).mean()
```

```python
x_values = train_dataset.iloc[:,[2, 4, 5, 6, 7, 9, 11]]
train_dataset.loc[ train_dataset['Fare'] <= 7.91, 'Fare'] = 0
train_dataset.loc[(train_dataset['Fare'] > 7.91) & (train_dataset['Fare'] <= 14.454), 'Fare'] = 1
train_dataset.loc[(train_dataset['Fare'] > 14.454) & (train_dataset['Fare'] <= 31), 'Fare']   = 2
train_dataset.loc[ train_dataset['Fare'] > 31, 'Fare'] = 3
train_dataset['Fare'] = train_dataset['Fare'].astype(int)
x_values.loc[x_values['Sex'] == 'male', 'Sex'] = 1
x_values.loc[x_values['Sex'] == 'female', 'Sex'] = 0
x_values['Sex'] = x_values['Sex'].astype(int)
x_values['Age'] = x_values['Age'].fillna(x_values['Age'].mean())
x_values['Embarked'] = x_values['Embarked'].map({'C': 0, 'Q': 1, 'S': 2, np.nan: 2}).astype(int)
```

```python
x_values.head()
```

```python
y_values = train_dataset.iloc[:, 1]
```

```python
testx = test_dataset.iloc[:, [1, 3, 4, 5, 6, 8, 10]]
testx['Fare'].fillna(testx['Fare'].dropna().median(), inplace=True)
testx.loc[ testx['Fare'] <= 7.91, 'Fare'] = 0
testx.loc[(testx['Fare'] > 7.91) & (testx['Fare'] <= 14.454), 'Fare'] = 1
testx.loc[(testx['Fare'] > 14.454) & (testx['Fare'] <= 31), 'Fare']   = 2
testx.loc[ testx['Fare'] > 31, 'Fare'] = 3
testx['Fare'] = testx['Fare'].astype(int)
```

```python
testx['Age'] = testx['Age'].fillna(testx['Age'].mean())
testx.loc[testx['Sex'] == 'male', 'Sex'] = 1
testx.loc[testx['Sex'] == 'female', 'Sex'] = 0
testx['Sex'] = testx['Sex'].astype(int)
testx['Embarked'] = testx['Embarked'].map({'C': 0, 'Q': 1, 'S': 2, np.nan: 2}).astype(int)
```

```python
testx.head()
```

```python
#Logistic Regression
logreg = LogisticRegression()
logreg.fit(x_values, y_values)
log_prediction = logreg.predict(testx)
acc_log = round(logreg.score(x_values, y_values) * 100, 2)

#k-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_values, y_values)
knn_prediction = knn.predict(testx)
acc_knn = round(knn.score(x_values, y_values) * 100, 2)

#Decision Trees
decision_tree = DecisionTreeClassifier()
decision_tree.fit(x_values, y_values)
dt_prediction = decision_tree.predict(testx)
acc_decision_tree = round(decision_tree.score(x_values, y_values) * 100, 2)

#Support Vector Machine
svc = SVC()
svc.fit(x_values, y_values)
svc_prediction = svc.predict(testx)
acc_svc = round(svc.score(x_values, y_values) * 100, 2)

#Naive Bayes
gaussian = GaussianNB()
gaussian.fit(x_values, y_values)
gaussian_prediction = gaussian.predict(testx)
acc_gaussian = round(gaussian.score(x_values, y_values) * 100, 2)
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression','Naive Bayes',
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_gaussian, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```

```python
df = pd.DataFrame()
df['PassengerId'] = test_dataset['PassengerId']
df['Survived'] = dt_prediction
```
